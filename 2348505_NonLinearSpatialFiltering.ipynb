**Maximum Filter**

```python
from PIL import Image, ImageFilter
```

**Maximum Filter**

---



Boldens Text

```python
im1 = Image.open("/content/minf.jpeg")

# applying the min filter
im2 = im1.filter(ImageFilter.MinFilter(size = 3))
display(im1)
display(im2)
```

**MaxFilter**


---


Thins Text

```python
im1 = Image.open("/content/maxf.jpeg")

# applying the max filter
im2 = im1.filter(ImageFilter.MaxFilter(size = 3))

display(im1)
display(im2)
```

**Median Filter**


---


Removes Noise

```python
im1 = Image.open("/content/noisy_img.jpeg")

# applying the max filter
im2 = im1.filter(ImageFilter.MedianFilter(size = 3))

display(im1)
display(im2)
```
