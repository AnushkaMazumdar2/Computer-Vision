**C) Using MatplotLib Library**

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# Reading the Images
img = mpimg.imread('img1.jpeg')

# Viewing the Images
plt.imshow(img)
```

```python
image = img[:, :, 0]
```

Using Various Filters

```python
plt.imshow(image, cmap="hot")
```

Using Various Interpolations

```python
plt.imshow(image, interpolation="bilinear")
```

Sometimes we want to enhance the contrast in our image, or expand the contrast in a particular region while sacrificing the detail in colors that don't vary much, or don't matter. A good tool to find interesting regions is the histogram.

```python
plt.hist(image.ravel(), bins=range(256), fc='k', ec='k')
```

**D) Using PIL Library**

The Python Imaging Library adds image processing capabilities to your Python interpreter. This library provides extensive file format support, an efficient internal representation, and fairly powerful image processing capabilities.

```python
from PIL import Image
```

```python
# Reading the Image
img = Image.open('img3.jpeg')

# Viewing the Image
display(img)
```

```python
# prints format of image
print(img.format)

# prints mode of image
# Eg- RGB, RFBA, CMYK, BGRetc.
print(img.mode)

# Prints the name of the file
print("Filename: ", img.filename)

# Prints the size as a width and height tuple (in pixels)
print("Size: ", img.size)

# Prints the width of the image (in pixels)
print("Width: ", img.width)

# Prints the height of the image (in pixels)
print("Height: ", img.height)

# Closing the image
img.close()
```

**ii) Converting the given images in to different formats (For ex: *.jpeg, *.jpg, *.jpe, , *.png, tiff, *.tif, e.t.c)**

We can use openCV library to write the images to different formats and then save it in the specified location

```python
#Converting the image to jpeg
img = cv2.imread("Flowers.png")
cv2.imwrite('flowers.jpeg',image)

#Converting the image to jpg
img = cv2.imread("Flowers.png")
cv2.imwrite('flower.jpg',image)

#Converting the image to jpe
img = cv2.imread("Flowers.png")
cv2.imwrite('flowers.jpe',image)

#Converting the image to png
img = cv2.imread("img3.jpeg")
cv2.imwrite('kittens.png',image)

#Converting the image to tiff
img = cv2.imread("img4.jpeg")
cv2.imwrite('cats.tiff',image)

#Converting the image to tif
img = cv2.imread("img2.jpeg")
cv2.imwrite('sky.tif',img)
```
