**Using FILTER2D FUNCTION**

Using this function, we can create a convolution between the image and the given kernel for creating filters like smoothing and blurring, sharpening, and edge detection in an image. This function will simply convolute the 2d matrix with the image at pixel level and produce an output image.

```python
# importing the modules needed
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# Reading and Viewing the Image
img = cv2.imread("img4.jpeg")
cv2_imshow(img)
```

```python
# Creating the kernel
kernel1 = np.ones((5, 5), np.float32)/30

# Applying the filter2D() function
img1 = cv2.filter2D(src=img, ddepth=-1, kernel=kernel1)

# Shoeing the original and output image
cv2_imshow(img1)

cv2.waitKey()
cv2.destroyAllWindows()
```

**Applying Various Kernels**

In image processing, a kernel, convolution matrix, or mask is a small matrix used for blurring, sharpening, embossing, edge detection, and more. This is accomplished by doing a convolution between the kernel and an image. Or more simply, when each pixel in the output image is a function of the nearby pixels (including itself) in the input image, the kernel is that function.

**i) Identity Kernel**

Identity kernel or the do nothing kernel basically displays the same image hence the name

```python
image = cv2.imread('img4.jpeg')

# Creating the Identity kernel(2d convolution matrix)
ikernel = np.array([[0, 0, 0],
                    [0, 1, 0],
                    [0, 0, 0]])

# Applying the filter2D() function
img = cv2.filter2D(src=image, ddepth=-1, kernel=ikernel)

# Shoeing the original and output image
cv2_imshow(image)
cv2_imshow(img)

cv2.waitKey()
cv2.destroyAllWindows()
```

**ii) Edge Detection Kernel**

Edge detection is an image-processing technique that is used to identify the boundaries (edges) of objects or regions within an image. Edges are among the most important features associated with images.

```python
image = cv2.imread('img4.jpeg')

# Creating the Edge Detection kernel(2d convolution matrix)
ikernel = np.array([[-1, -1, -1],
                    [-1, 8, -1],
                    [-1, -1, -1]])

# Applying the filter2D() function
img = cv2.filter2D(src=image, ddepth=-1, kernel=ikernel)

# Shoeing the original and output image
cv2_imshow(image)
cv2_imshow(img)

cv2.waitKey()
cv2.destroyAllWindows()
```

**iii) Sharpen Kernel**

The sharpen kernel emphasizes differences in adjacent pixel values. This makes the image look more vivid.

```python
image = cv2.imread('img4.jpeg')

# Creating the Sharpen kernel(2d convolution matrix)
ikernel = np.array([[0, -1, 0],
                    [-1, 5, -1],
                    [0, -1, 0]])

# Applying the filter2D() function
img = cv2.filter2D(src=image, ddepth=-1, kernel=ikernel)

# Shoeing the original and output image
cv2_imshow(image)
cv2_imshow(img)

cv2.waitKey()
cv2.destroyAllWindows()
```

**iv) BoxBlur Kernel**

A box blur (also known as a box linear filter) is a spatial domain linear filter in which each pixel in the resulting image has a value equal to the average value of its neighboring pixels in the input image. It is a form of low-pass ("blurring") filter

```python
image = cv2.imread('img4.jpeg')

# Creating the BoxBlur kernel(2d convolution matrix)
ikernel = np.array([[1/9, 1/9, 1/9],
                    [1/9, 1/9, 1/9],
                    [1/9, 1/9, 1/9]])

# Applying the filter2D() function
img = cv2.filter2D(src=image, ddepth=-1, kernel=ikernel)

# Shoeing the original and output image
cv2_imshow(image)
cv2_imshow(img)

cv2.waitKey()
cv2.destroyAllWindows()
```

**v) Gaussian Blur Kernel**

It is a widely used effect in graphics software, typically to reduce image noise and reduce detail. The visual effect of this blurring technique is a smooth blur resembling that of viewing the image through a translucent screen, distinctly different from the bokeh effect produced by an out-of-focus lens or the shadow of an object under usual illumination.

```python
image = cv2.imread('img4.jpeg')

# Creating the Gaussian Blur kernel(2d convolution matrix)
ikernel = np.array([[1/256, 4/256, 6/256,4/256,1/256],
                    [4/256, 16/256, 24/256,16/256,4/256],
                    [6/256, 24/256, 36/256,24/256,6/256],
                    [4/256, 16/256, 24/256,16/256,4/256],
                    [1/256, 4/256, 6/256,4/256,1/256]])

# Applying the filter2D() function
img = cv2.filter2D(src=image, ddepth=-1, kernel=ikernel)

# Shoeing the original and output image
cv2_imshow(image)
cv2_imshow(img)

cv2.waitKey()
cv2.destroyAllWindows()
```

**Applying Convolve Function**

```python
from scipy import ndimage
```

Using Reflect Mode:

The input is extended by reflecting about the edge of the last pixel. This mode is also sometimes referred to as half-sample symmetric.

```python
a = np.array([[1, 0, 3, 5],
              [9, 5, 1, 8],
              [3, 3, 6, 7],
              [5, 1, 8, 0]])
k = np.array([[1,1,1],[1,1,0],[1,0,0]])
ndimage.convolve(a, k, mode='reflect')
```

Using Constant Mode:


The input is extended by filling all values beyond the edge with the same constant value, defined by the cval parameter.

```python
a = np.array([[1, 0, 3, 5],
              [9, 5, 1, 8],
              [3, 3, 6, 7],
              [5, 1, 8, 0]])
k = np.array([[1,1,1],[1,1,0],[1,0,0]])
ndimage.convolve(a, k, mode='constant', cval=0.4)
```

Using Nearest Mode:

The input is extended by replicating the last pixel.

```python
a = np.array([[1, 0, 3, 5],
              [9, 5, 1, 8],
              [3, 3, 6, 7],
              [5, 1, 8, 0]])
k = np.array([[1,1,1],[1,1,0],[1,0,0]])
ndimage.convolve(a, k, mode='nearest')
```

Using Mirror Mode:

The input is extended by reflecting about the center of the last pixel. This mode is also sometimes referred to as whole-sample symmetric.

```python
a = np.array([[1, 0, 3, 5],
              [9, 5, 1, 8],
              [3, 3, 6, 7],
              [5, 1, 8, 0]])
k = np.array([[1,1,1],[1,1,0],[1,0,0]])
ndimage.convolve(a, k, mode='mirror')
```

Using Wrap Mode:

The input is extended by wrapping around to the opposite edge.

```python
a = np.array([[1, 0, 3, 5],
              [9, 5, 1, 8],
              [3, 3, 6, 7],
              [5, 1, 8, 0]])
k = np.array([[1,1,1],[1,1,0],[1,0,0]])
ndimage.convolve(a, k, mode='wrap')
```
