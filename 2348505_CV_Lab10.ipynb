**Importing Necessary Libraries**

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import random
```

# **Demonstrate frame extraction from the video and display the color components of the images**

**1. Extracting all the frames from the video**

```python
cap = cv2.VideoCapture('/content/videoplayback.mp4')
frames = []
if cap.isOpened():
    current_frame = 0
    while True:
        ret, frame = cap.read()
        if ret:
            name = f'frameIn/frame{current_frame}.jpg'
            print(f"Creating file... {name}")
            cv2.imwrite(name, frame)
            frames.append(name)
        current_frame += 1
    cap.release()
cv2.destroyAllWindows()
```

```python
# Visualising a random frame saved from the video
rn = random.randint(1, 1416)
frame = cv2.imread(f"frame{rn}.jpg")
cv2_imshow(frame)
```

```python
# Visualising a random frame saved from the video
rn = random.randint(1, 1416)
frame = cv2.imread(f"frame{rn}.jpg")
cv2_imshow(frame)
```

```python
# Extracting color components of the image

b,g,r = cv2.split(frame)

plt.figure(figsize=(15,10))
plt.subplot(131)
plt.title("Blue Component")
plt.imshow(b,"gray")
plt.axis("off")

plt.subplot(132)
plt.title("Green Component")
plt.imshow(g,"gray")
plt.axis("off")

plt.subplot(133)
plt.title("Red Component")
plt.imshow(r,"gray")
plt.axis("off")
```

**2. Extracting specific/desired frames from the video**

```python
video = cv2.VideoCapture('/content/videoplayback.mp4')
```

```python
fps = video.get(cv2.CAP_PROP_FPS)
print('frames per second =',fps)
```

```python
# Extracting 700th frame 28 seconds into the video
minutes = 0
seconds = 28
frame_id = int(fps*(minutes*60 + seconds))
print('frame id =',frame_id)
```

```python
video.set(cv2.CAP_PROP_POS_FRAMES, frame_id)
ret, frame = video.read()
```

```python
cv2_imshow( frame); cv2.waitKey(0)
cv2.imwrite('frame_671.png', frame)
```

```python
# Extracting color components of the image
frame = cv2.imread("frame_671.png")

b,g,r = cv2.split(frame)

plt.figure(figsize=(15,10))
plt.subplot(131)
plt.title("Blue Component")
plt.imshow(b,"gray")
plt.axis("off")

plt.subplot(132)
plt.title("Green Component")
plt.imshow(g,"gray")
plt.axis("off")

plt.subplot(133)
plt.title("Red Component")
plt.imshow(r,"gray")
plt.axis("off")
```
